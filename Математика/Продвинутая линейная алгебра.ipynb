```python
"""Уровень 0:
Задание 1
Дан объект в пространстве, при помощи линейных отображений:
Уменьшить объект в два раза
Повернуть на 130 градусов
Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])"""

import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
"""Уменьшить объект в два раза"""

T = A/2

A = np.array(T)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
"""Повернуть на 130 градусов"""
"""создаем матрицу вращения"""

rot = np.array([
    [np.cos(130), -np.sin(130),0],
    [np.sin(130), np.cos(130),0],
    [0,0,1]
    ])

"""переумножаем матрицы с транспонированием"""
res = rot @ A.transpose()
T1 = res.transpose()

A = np.array(T1)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
"""Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])"""
"""Мaтрица отражения"""
matrix = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
    ])

"""переумножаем матрицы с транспонированием"""
res2 = matrix @ A.transpose()
T2 = res2.transpose()

A = np.array(T2)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```
